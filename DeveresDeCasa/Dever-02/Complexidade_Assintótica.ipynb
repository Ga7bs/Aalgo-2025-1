```python
import time
import sys

def fatorial_iterativo(n):
    """Calcula o fatorial de n de forma iterativa."""
    if n < 0:
        raise ValueError("Fatorial não é definido para números negativos.")
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i
    return resultado

def medir_tempo_execucao(n):
    """Mede e retorna o tempo de execução da função de fatorial para um dado n."""
    start_time = time.time()  # Inicia o cronômetro
    result = fatorial_iterativo(n)  # Chama a função fatorial
    end_time = time.time()    # Para o cronômetro
    return result, end_time - start_time  # Retorna o resultado e o tempo decorrido

# Testando o tempo de execução para diferentes valores de n
valores_n = [10, 100, 500, 1000]
resultados = {}

for n in valores_n:
    resultado, tempo = medir_tempo_execucao(n)
    resultados[n] = (resultado, tempo)

# Exibindo os resultados
for n, (resultado, tempo) in resultados.items():
    print(f"Fatorial de {n} é: {resultado}")
    print(f"Tempo de execução para n={n}: {tempo:.10f} segundos")

```
